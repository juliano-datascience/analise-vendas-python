```python
import pandas as pd
from google.colab import drive

# montar drive
drive.mount('/content/drive')

# caminho
caminho = "/content/drive/MyDrive/PROJETO VENDAS/Vendas+-+Base+de+Dados.xlsx"

# carregar
df = pd.read_excel(caminho)

# limpar colunas
df.columns = df.columns.str.strip()

print(df.head())
```

```python
df["Faturamento"] = df["Quantidade"] * df["Valor Unitário"]

print(df.head())
```

```python
top_produtos = (
    df.groupby("Produto")["Quantidade"]
    .sum()
    .sort_values(ascending=False)
)

print(top_produtos.head(10))
```

```python
faturamento = (
    df.groupby("Produto")["Faturamento"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

print(faturamento.head(10))
```

```python
total = faturamento["Faturamento"].sum()

faturamento["% Participação"] = (
    faturamento["Faturamento"] / total * 100
)

print(faturamento.head(10))
```

```python
faturamento["% Acumulado"] = faturamento["% Participação"].cumsum()

def classificar_abc(valor):
    if valor <= 80:
        return "A"
    elif valor <= 95:
        return "B"
    else:
        return "C"

faturamento["Classe ABC"] = faturamento["% Acumulado"].apply(classificar_abc)

print(faturamento.head(10))
```

```python
import plotly.express as px

fig = px.bar(
    faturamento,
    x="Produto",
    y="Faturamento",
    color="Classe ABC",
    title="Curva ABC de Produtos",
    text_auto=True
)

fig.update_layout(xaxis_tickangle=-45)

fig.show()
```

```python
preco_medio = (
    df.groupby("Produto")["Valor Unitário"]
    .mean()
    .sort_values(ascending=False)
)

print(preco_medio.head(10))
```
